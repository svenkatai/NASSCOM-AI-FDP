#1. Derivatives — the slope of a function

```python
# -----------------------------------------------------------
# 🔹 1A. NUMERICAL DERIVATIVE (finite difference)
# -----------------------------------------------------------

# The derivative is the slope: how much f changes for a tiny step h
def f(x):
    return x**2

h = 1e-6
slope_at_3 = (f(3 + h) - f(3)) / h
print('Numerical f\'(3):', round(slope_at_3, 4), ' (exact = 6)')
```

```python
# -----------------------------------------------------------
# 🔹 1B. SYMBOLIC DERIVATIVE (SymPy)
# -----------------------------------------------------------

expr = x**2
deriv = sp.diff(expr, x)          # differentiate w.r.t. x
print('d/dx (x**2) =', deriv)     # 2*x

# Evaluate the symbolic derivative at x = 3
print('Symbolic f\'(3) =', deriv.subs(x, 3))
```

#### 🧪 LAB EXERCISE 1 — Numerical vs symbolic derivative

For the function `g(x) = x**3 + 2*x`:
1. Compute the **numerical** derivative at `x = 2` using a finite difference.
2. Compute the **symbolic** derivative with `sp.diff` and print it.
3. Evaluate the symbolic derivative at `x = 2` and confirm it matches step 1.

```python
def g(x):
    return x**3 + 2*x

# 1. Numerical derivative at x = 2 (use h = 1e-6)
h = 1e-6
numerical_deriv_at_2 = (g(2 + h) - g(2)) / h
print('Numerical g\'(2):', round(numerical_deriv_at_2, 4))

# 2. Symbolic derivative of x**3 + 2*x
expr_g = x**3 + 2*x
symbolic_deriv = sp.diff(expr_g, x)
print('d/dx (x**3 + 2*x) =', symbolic_deriv)

# 3. Evaluate the symbolic derivative at x = 2 and compare
symbolic_deriv_at_2 = symbolic_deriv.subs(x, 2)
print('Symbolic g\'(2) =', symbolic_deriv_at_2)
# Use np.isclose for a robust comparison between numerical and symbolic results
print('Numerical and Symbolic match:', np.isclose(numerical_deriv_at_2, float(symbolic_deriv_at_2), atol=1e-4))
```

#2. Partial derivatives & the gradient

```python
# -----------------------------------------------------------
# 🔹 2A. PARTIAL DERIVATIVES
# -----------------------------------------------------------

f2 = x**2 + 3*x*y + y**2

# A partial derivative differentiates ONE variable, holding others fixed
print('df/dx =', sp.diff(f2, x))     # 2*x + 3*y
print('df/dy =', sp.diff(f2, y))     # 3*x + 2*y
```

```python
# -----------------------------------------------------------
# 🔹 2B. THE GRADIENT (vector of partials)
# -----------------------------------------------------------

# The gradient stacks every partial derivative into one vector
grad = [sp.diff(f2, v) for v in (x, y)]
print('grad f =', grad)

# Evaluate the gradient at the point (x=1, y=2)
grad_at = [g.subs({x: 1, y: 2}) for g in grad]
print('grad f at (1, 2) =', grad_at)   # points in the steepest-ascent direction
```

#### 🧪 LAB EXERCISE 2 — Compute a gradient

For `h2 = x**2 * y + sp.sin(y)`:
1. Compute `dh/dx` and `dh/dy` with `sp.diff`.
2. Assemble the gradient as a list `[dh/dx, dh/dy]`.
3. Evaluate the gradient at the point `(x=2, y=0)`.

```python
h2 = x**2 * y + sp.sin(y)

# 1. dh/dx and dh/dy
dh_dx = sp.diff(h2, x)
dh_dy = sp.diff(h2, y)
print('dh/dx =', dh_dx)
print('dh/dy =', dh_dy)

# 2. Assemble the gradient list
gradient_h2 = [dh_dx, dh_dy]
print('Gradient h2 =', gradient_h2)

# 3. Evaluate at (x=2, y=0)  -> hint: .subs({x: 2, y: 0})
gradient_at_point = [g.subs({x: 2, y: 0}) for g in gradient_h2]
print('Gradient at (2, 0) =', gradient_at_point)
```

#3. The chain rule

```python
# -----------------------------------------------------------
# 🔹 3A. CHAIN RULE BY HAND vs SymPy
# -----------------------------------------------------------

# y = sin(x**2) is a composition: outer = sin(u), inner = u = x**2
# Chain rule:  dy/dx = cos(u) * du/dx = cos(x**2) * 2x
by_hand = sp.cos(x**2) * 2*x
by_sympy = sp.diff(sp.sin(x**2), x)

print('By hand :', by_hand)
print('By SymPy:', by_sympy)
print('Match?  ', sp.simplify(by_hand - by_sympy) == 0)
```

```python
# -----------------------------------------------------------
# 🔹 3B. CHAINING THREE FUNCTIONS
# -----------------------------------------------------------

# y = (3x + 1)**4  -> outer^4, inner (3x+1)
expr3 = (3*x + 1)**4
print('d/dx (3x+1)^4 =', sp.diff(expr3, x))   # 12*(3x+1)^3
```

#4. Backpropagation — the chain rule in a network

A 2-layer network is just a composition:  **x → (W1) → ReLU → (W2) → y_hat**. Backprop applies the chain rule from the loss backwards to every weight.

```python
# -----------------------------------------------------------
# 🔹 4A. FORWARD PASS
# -----------------------------------------------------------

# Tiny toy problem: 4 samples, 3 input features, 5 hidden units, 1 output
X = np.random.randn(4, 3)
Y = np.random.randn(4, 1)
W1 = np.random.randn(3, 5) * 0.1
W2 = np.random.randn(5, 1) * 0.1

z1 = X @ W1                 # linear layer 1
h  = np.maximum(0, z1)      # ReLU activation
y_hat = h @ W2              # linear layer 2 (prediction)
loss = ((y_hat - Y) ** 2).mean()
print('Initial loss:', round(loss, 4))
```

```python
# -----------------------------------------------------------
# 🔹 4B. BACKWARD PASS (gradients via the chain rule)
# -----------------------------------------------------------

# Work backwards from the loss, one link at a time
dy   = 2 * (y_hat - Y) / Y.size      # d loss / d y_hat
dW2  = h.T @ dy                      # d loss / d W2
dh   = dy @ W2.T                     # d loss / d h
dz1  = dh * (z1 > 0)                 # ReLU gradient (1 where z1>0 else 0)
dW1  = X.T @ dz1                     # d loss / d W1

print('dW1 shape:', dW1.shape, '(matches W1)')
print('dW2 shape:', dW2.shape, '(matches W2)')
```

```python
# -----------------------------------------------------------
# 🔹 4C. ONE GRADIENT-DESCENT STEP SHOULD LOWER THE LOSS
# -----------------------------------------------------------

lr = 0.1
W1 -= lr * dW1               # step downhill
W2 -= lr * dW2

# Recompute the loss after the update
h_new = np.maximum(0, X @ W1)
loss_new = ((h_new @ W2 - Y) ** 2).mean()
print('Loss before:', round(loss, 4))
print('Loss after :', round(loss_new, 4), '-> should be lower')
```

#### 🧪 LAB EXERCISE 4 — Derive the gradients for a 2-layer network

Fresh weights are set up below. Using the chain rule (reuse the pattern from 4A/4B):
1. Run the **forward pass**: compute `z1`, `h` (ReLU), `y_hat`, and `loss`.
2. Run the **backward pass**: compute `dy`, `dW2`, `dh`, `dz1`, `dW1`.
3. Take one gradient-descent step (`lr = 0.05`) and print the loss before and after.

```python
Xb = np.random.randn(6, 4)
Yb = np.random.randn(6, 1)
Wa = np.random.randn(4, 8) * 0.1     # input -> hidden
Wb = np.random.randn(8, 1) * 0.1     # hidden -> output

# 1. Forward pass: z1, h = ReLU(z1), y_hat, loss
z1 = Xb @ Wa
h = np.maximum(0, z1)
y_hat = h @ Wb
loss = ((y_hat - Yb) ** 2).mean()
print('Initial loss:', round(loss, 4))

# 2. Backward pass: dy, dWb, dh, dz1, dWa
dy = 2 * (y_hat - Yb) / Yb.size
dWb = h.T @ dy
dh = dy @ Wb.T
dz1 = dh * (z1 > 0)
dWa = Xb.T @ dz1

# 3. One step with lr = 0.05; print loss before and after
lr = 0.05
Wa -= lr * dWa
Wb -= lr * dWb

# Recompute the loss after the update
h_new = np.maximum(0, Xb @ Wa)
loss_new = ((h_new @ Wb - Yb) ** 2).mean()
print('Loss before:', round(loss, 4))
print('Loss after :', round(loss_new, 4), '-> should be lower')
```

#5. Hessian & gradient descent

```python
# -----------------------------------------------------------
# 🔹 5A. THE HESSIAN (matrix of second derivatives = curvature)
# -----------------------------------------------------------

f5 = x**2 + 3*x*y + y**2
H = sp.hessian(f5, (x, y))
print('Hessian of f:')
sp.pprint(H)        # [[2, 3], [3, 2]]
```

```python
# -----------------------------------------------------------
# 🔹 5B. GRADIENT DESCENT ON A SIMPLE FUNCTION
# -----------------------------------------------------------

# Minimise f(x) = (x - 4)**2, whose minimum is at x = 4.
# Update rule:  x <- x - lr * f'(x),  with f'(x) = 2*(x - 4)
xv = 0.0          # starting point
lr = 0.2
for step in range(15):
    grad = 2 * (xv - 4)        # the derivative
    xv = xv - lr * grad        # step against the gradient
print('Converged x:', round(xv, 3), ' (true minimum = 4)')
```

| ------- | ------------- | --------- |
| **Derivative** | slope / rate of change | finite difference · `sp.diff` |
| **Partial derivative** | slope along one variable | `sp.diff(f, x)` |
| **Gradient** | vector of all partials → steepest ascent | `[sp.diff(f, v) for v in vars]` |

| **Hessian** | second derivatives → curvature | `sp.hessian(f, vars)` |
| **Gradient descent** | step downhill | `x -= lr * grad` |

### Computing the Gradient Vector ∇f

The gradient vector $\nabla f$ (pronounced "nabla f") for a multivariate function $f(x_1, x_2, ..., x_n)$ is a vector whose components are the partial derivatives of $f$ with respect to each variable. It points in the direction of the steepest ascent of the function.

For a function $f(x, y)$, the gradient is defined as:

$\nabla f(x, y) = \left[ \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right]$

### Let's use a new multivariate function, for example: f_new = x**3 + 2*y**2 - 5*x*y
f_new = x**3 + 2*y**2 - 5*x*y

# 1. Compute the partial derivatives with respect to x and y
df_dx = sp.diff(f_new, x)
df_dy = sp.diff(f_new, y)

print(f'Partial derivative with respect to x (df/dx): {df_dx}')
print(f'Partial derivative with respect to y (df/dy): {df_dy}')

# 2. Assemble the gradient vector
gradient_f_new = [df_dx, df_dy]
print(f'\nGradient vector \u2207f_new: {gradient_f_new}')

# 3. Evaluate the gradient at a specific point, e.g., (x=1, y=3)
point = {x: 1, y: 3}
gradient_at_point = [g.subs(point) for g in gradient_f_new]
print(f'\nGradient \u2207f_new at (1, 3): {gradient_at_point}')

# The result [3*(1)**2 - 5*3, 4*3 - 5*1] = [3 - 15, 12 - 5] = [-12, 7]
# This vector indicates that at (1,3), the function f_new increases most rapidly in the direction of (-12, 7).

### Visualizing the Gradient Vector

The gradient vector points in the direction of the steepest ascent of the function. Let's visualize this for our function $f_{new} = x^3 + 2y^2 - 5xy$ at the point $(1, 3)$.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the numerical version of the function for plotting
def f_new_numeric(x_val, y_val):
    return x_val**3 + 2*y_val**2 - 5*x_val*y_val

# Create a grid for x and y values
x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)
X_grid, Y_grid = np.meshgrid(x_vals, y_vals)
Z = f_new_numeric(X_grid, Y_grid)

# Define the point (x0, y0) and the gradient vector (grad_x, grad_y) at that point
x0, y0 = 1, 3
# From previous execution, gradient at (1,3) is [-12, 7]
grad_x, grad_y = -12, 7

plt.figure(figsize=(8, 7))

# Plot the contour lines of the function
plt.contour(X_grid, Y_grid, Z, levels=20, cmap='viridis')
plt.colorbar(label='f_new(x, y)')

# Plot the specific point (x0, y0)
plt.plot(x0, y0, 'ro', markersize=8, label=f'Point ({x0}, {y0})')

# Plot the gradient vector as an arrow starting from (x0, y0)
# The 'scale' parameter controls the length of the arrow relative to its components
plt.quiver(x0, y0, grad_x, grad_y, color='red', scale=50, linewidth=1, label=f'Gradient (dx={grad_x}, dy={grad_y})')

plt.title('Contour Plot of f_new(x, y) with Gradient Vector at (1, 3)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

### Computing the Hessian Matrix

The Hessian matrix is a square matrix of second-order partial derivatives of a scalar-valued function. It describes the local curvature of a function. For a function $f(x, y)$, the Hessian matrix $H$ is:

$H(f) = \begin{pmatrix}
\frac{\partial^2 f}{\partial x^2} & \frac{\partial^2 f}{\partial x \partial y} \\
\frac{\partial^2 f}{\partial y \partial x} & \frac{\partial^2 f}{\partial y^2}
\end{pmatrix}$

```python
# We'll use the same f_new function: f_new = x**3 + 2*y**2 - 5*x*y
# from our gradient calculation

# Compute the Hessian of f_new with respect to (x, y)
Hessian_f_new = sp.hessian(f_new, (x, y))

print('Hessian of f_new:')
sp.pprint(Hessian_f_new)
```

### Calculating Eigenvalues of the Hessian at a Specific Point

Eigenvalues of the Hessian matrix provide information about the nature of a critical point (minimum, maximum, or saddle point). We'll evaluate the Hessian at `(1, 3)` and then compute its eigenvalues.

```python
# Recall the Hessian of f_new:
# Hessian_f_new = sp.Matrix([[6*x, -5], [-5, 4]])

# Substitute x=1 and y=3 into the Hessian matrix
point_for_hessian = {x: 1, y: 3}
numerical_hessian = Hessian_f_new.subs(point_for_hessian)
print('Numerical Hessian at (1, 3):')
sp.pprint(numerical_hessian)

# Calculate the eigenvalues of the numerical Hessian
eigenvalues = numerical_hessian.eigenvals()
print('\nEigenvalues of the Hessian at (1, 3):')
print(eigenvalues)

# The keys of the dictionary are the eigenvalues, and the values are their multiplicities.
# For example, {lambda1: multiplicity1, lambda2: multiplicity2}
```

### Visualizing the Saddle Point in 3D

To better understand why `(1, 3)` is a saddle point, we can visualize the function `f_new` in 3D. A saddle point is characterized by a surface that curves upwards in some directions and downwards in others, resembling a saddle.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Define the numerical version of the function for plotting
def f_new_numeric(x_val, y_val):
    return x_val**3 + 2*y_val**2 - 5*x_val*y_val

# Create a grid for x and y values around the saddle point (1, 3)
x_range = np.linspace(-2, 4, 100)
y_range = np.linspace(0, 6, 100)
X_grid, Y_grid = np.meshgrid(x_range, y_range)
Z_grid = f_new_numeric(X_grid, Y_grid)

# Create the 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X_grid, Y_grid, Z_grid, cmap='viridis', alpha=0.8)

# Mark the saddle point (1, 3)
x_saddle, y_saddle = 1, 3
z_saddle = f_new_numeric(x_saddle, y_saddle)
ax.scatter(x_saddle, y_saddle, z_saddle, color='red', s=100, label=f'Saddle Point ({x_saddle}, {y_saddle})', depthshade=True)

# Add labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('f_new(X, Y)')
ax.set_title('3D Plot of f_new(x, y) with Saddle Point at (1, 3)')
ax.legend()
plt.show()
```
