```python
# ==========================================================
# Hands-On #1: Vectors, Dot Products & Norms in NumPy
# ==========================================================

import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# ==========================================================
# 1. Creating Vectors & Basic Operations
# ==========================================================

print("\n========== 1. VECTOR OPERATIONS ==========\n")

u = np.array([2, 3])
v = np.array([4, 1])

print("Vector u:", u)
print("Vector v:", v)

# Vector Addition
print("\nu + v =", u + v)

# Scalar Multiplication
print("3 * u =", 3 * u)

# Vector Subtraction
print("u - v =", u - v)

# Element-wise Multiplication
print("u * v =", u * v)

# ==========================================================
# 2. Dot Product & Angle Between Vectors
# ==========================================================

print("\n========== 2. DOT PRODUCT ==========\n")

# Method 1: np.dot()
dot1 = np.dot(u, v)

# Method 2: @ Operator
dot2 = u @ v

print("Dot Product using np.dot() =", dot1)
print("Dot Product using @ operator =", dot2)

# Compute Cosine of Angle
cos_theta = np.dot(u, v) / (
    np.linalg.norm(u) * np.linalg.norm(v)
)

# Angle in Radians
theta_rad = np.arccos(cos_theta)

# Convert to Degrees
theta_deg = np.degrees(theta_rad)

print("\nCos(theta) =", cos_theta)
print("Angle between vectors = {:.2f} degrees".format(theta_deg))

# ==========================================================
# 3. Norms of a Vector
# ==========================================================

print("\n========== 3. VECTOR NORMS ==========\n")

w = np.array([3, -4])

print("Vector w:", w)

# L1 Norm (Manhattan Distance)
l1 = np.linalg.norm(w, 1)

# L2 Norm (Euclidean Distance)
l2 = np.linalg.norm(w, 2)

# Infinity Norm
linf = np.linalg.norm(w, np.inf)

print("L1 Norm   =", l1)
print("L2 Norm   =", l2)
print("L∞ Norm   =", linf)

# ==========================================================
# 4. Cosine Similarity Example
# ==========================================================

print("\n========== 4. COSINE SIMILARITY ==========\n")

a = np.array([[1, 2, 3]])
b = np.array([[2, 4, 6]])

similarity = cosine_similarity(a, b)

print("Vector A:", a)
print("Vector B:", b)
print("Cosine Similarity =", similarity)

# ==========================================================
# 5. Exercise: Student Exam Scores Analysis
# ==========================================================

print("\n========== 5. STUDENT PERFORMANCE ANALYSIS ==========\n")

# Scores = [Math, Science, English]

student1 = np.array([85, 90, 88])
student2 = np.array([80, 85, 84])
student3 = np.array([92, 70, 65])

students = {
    "Student 1": student1,
    "Student 2": student2,
    "Student 3": student3
}

# ----------------------------------------------------------
# L2 Norms (Overall Performance)
# ----------------------------------------------------------

print("----- L2 Norms (Overall Performance) -----")

for name, scores in students.items():
    norm = np.linalg.norm(scores)
    print(f"{name}: {norm:.2f}")

# ----------------------------------------------------------
# Cosine Similarities
# ----------------------------------------------------------

print("\n----- Cosine Similarities -----")

sim_12 = cosine_similarity([student1], [student2])[0][0]
sim_13 = cosine_similarity([student1], [student3])[0][0]
sim_23 = cosine_similarity([student2], [student3])[0][0]

print(f"Student1 vs Student2 = {sim_12:.4f}")
print(f"Student1 vs Student3 = {sim_13:.4f}")
print(f"Student2 vs Student3 = {sim_23:.4f}")

# ----------------------------------------------------------
# Interpretation
# ----------------------------------------------------------

print("\n----- Interpretation -----")

highest_norm_student = max(
    students,
    key=lambda x: np.linalg.norm(students[x])
)

print(f"Highest Overall Performance: {highest_norm_student}")

if sim_12 > sim_13 and sim_12 > sim_23:
    print("Student 1 and Student 2 have the most similar score pattern.")
elif sim_13 > sim_23:
    print("Student 1 and Student 3 have the most similar score pattern.")
else:
    print("Student 2 and Student 3 have the most similar score pattern.")

print("\nAnalysis Complete!")
```
