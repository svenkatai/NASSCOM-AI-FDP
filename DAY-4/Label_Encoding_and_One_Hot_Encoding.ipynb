#Label Encoding and One-Hot Encoding

Both are categorical encoding techniques used to convert non-numeric (categorical) data into a numerical format that machine learning models can understand.

#🔹 1. Label Encoding

📌 Definition:

Converts each category into a unique integer.

✅ Use Case:

When categorical values have a natural ordinal relationship (e.g., "Low" < "Medium" < "High")

🔄 Example:

```python
from sklearn.preprocessing import LabelEncoder

data = ['Red', 'Green', 'Blue', 'Red']
le = LabelEncoder()
encoded = le.fit_transform(data)
print(encoded)  # Output: [2 1 0 2]
```

| Category | Encoded Value |
| -------- | ------------- |
| Blue     | 0             |
| Green    | 1             |
| Red      | 2             |


#🔹 2. One-Hot Encoding
📌 Definition:

Converts each category into a binary vector (0s and 1s), with one hot (1) position indicating the category.

✅ Use Case:

Best for nominal data (e.g., 'Red', 'Green', 'Blue')

🔄 Example:

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

data = np.array([['Red'], ['Green'], ['Blue'], ['Red']])
ohe = OneHotEncoder()
encoded = ohe.fit_transform(data)
print(encoded)
```

| Red | Green | Blue |
| --- | ----- | ---- |
| 1   | 0     | 0    |
| 0   | 1     | 0    |
| 0   | 0     | 1    |
| 1   | 0     | 0    |


#Summary Table:

| Feature                   | Label Encoding       | One-Hot Encoding    |
| ------------------------- | -------------------- | ------------------- |
| Output Type               | Integer values       | Binary matrix       |
| Suitable For              | Ordinal categorical  | Nominal categorical |
| Introduces Order?         | ✅ Yes                | ❌ No                |
| Memory Efficient?         | ✅ Yes                | ❌ No (more columns) |
| Risk of Misleading Model? | ✅ Yes (if unordered) | ❌ No                |
